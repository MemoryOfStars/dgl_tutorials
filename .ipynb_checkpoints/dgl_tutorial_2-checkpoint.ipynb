#### 使用add_nodes()来一个个加入节点，适用于那些图不能一次性构建完的应用

```python
g = dgl.DGLGraph()
g.add_nodes(10)

for i in range(1,4):
    g.add_edge(i, 0)

src = list(range(5, 8));dst = [0]*3
g.add_edges(src, dst)
src = th.tensor([8,9]);dst = th.tensor([0,0])
g.add_edges(src, dst)

g = dgl.DGLGraph()
g.add_nodes(10)
src = th.tensor(list(range(1,10)));
g.add_edges(src, 0)

nx.draw(g.to_networkx(), with_labels=True)
plt.show()
```

#### 给节点和边附上feature值 这个ndata是[string : data]构成的，这个字典叫做field

```python
import dgl
import torch as th

x = th.randn(10, 3)
g.ndata['x'] = x
g.ndata
```

```python
g.ndata['x'][0] = th.zeros(1,3)
g.ndata['x'][[0,1,2]] = th.zeros(3,3)
g.ndata['x'][th.tensor([0,1,2])] = th.randn((3,3))
g.ndata
```

#### edata是边的feature

```python
g.edata['w'] = th.randn(9,2)

g.edata['w'][1] = th.randn(1,2)
g.edata['w'][[0,1,2]] = th.zeros(3,2)
g.edata['w'][th.tensor([0,1,2])] = th.zeros(3,2)

g.edata['w'][g.edge_id(1,0)] = th.ones(1,2)
g.edata['w'][g.edge_ids([1,2,3], [0,0,0])] = th.ones(3,2) #使用边的两个端点来确定一个有向边
g.edata['w'][g.edge_ids([1,2,3], [0,0,0])] = th.ones(3,2)
g.edata
```

```python
print(g.node_attr_schemes()) # 赋值之后就会自动检测dim，生成一个attr_scheme，包含了feature的shape信息
g.ndata['x'] = th.zeros((10, 4))
print(g.node_attr_schemes())
```

#### 使用多重图（两个点有多个边）

```python
g_multi = dgl.DGLGraph()
g_multi.add_nodes(10)
g_multi.ndata['x'] = th.randn(10, 2)

g_multi.add_edges(list(range(1, 10)), 0)
g_multi.add_edge(1, 0)

g_multi.edata['w'] = th.randn(10, 2)
g_multi.edges[1].data['w'] = th.zeros(1,2)
print(g_multi.edges())
```

#### 多重图中的边不能通过起点终点来唯一确定，要使用$edge\_id$来获取edge的ID
